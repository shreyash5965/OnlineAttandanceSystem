```python
import numpy as np
import cv2
import insightface
from insightface.app import FaceAnalysis
from insightface.data import get_image as ins_get_image
```

# Insightface Configuration

```python
# model_pack_name = 'buffalo_sc'
app_l = FaceAnalysis(name='buffalo_l',providers=['CPUExecutionProvider'])
app_l.prepare(ctx_id=0,det_size=(640,640))
```

```python
print(app_l)
```

```python
app_sc = FaceAnalysis(name='buffalo_sc',providers=['CPUExecutionProvider'])
app_sc.prepare(ctx_id=0,det_size=(640,640))
```

# Load Image

```python
img = cv2.imread('Images/532.jpg')
# cv2.imshow('Initial Image',img)
# image window will close as soon as esc is pressed or X button is clicked
# cv2.waitKey(0)
# cv2.destroyAllWindows()
# print(img)
```

# Apply Image to Insightface Model

```python
result_l = app_l.get(img)
```

```python
print(result_l[0])
```

```python
# length of result gives the number of faces detected in the image
len(result_l)
```

```python
# keys of which value are printed in result_l
result_l[0].keys()
```

```python
# bbox value returns the co-ordinate of detected face x1,y1,x2,y2
result_l[0]['bbox']
```

```python
# gender returned by model is 0 for female and 1 for male
# gender_data has the respective gender in their index which is used to display gender in image
gender_data = ['Female','Male']
image_v2 = img.copy()
# draw rectangle to show detected face using bbox
for points in result_l:
    # convert bbox value to int as image accepts int only
    x1,y1,x2,y2 = points['bbox'].astype(int)
    # draw rectangle around detected face
    cv2.rectangle(image_v2,(x1,y1),(x2,y2),(0,255,0),2)
    
    # key points
    kp = points['kps'].astype(int)
    for k1,k2 in kp:
        cv2.circle(image_v2,(k1,k2),2,(0,255,255),-1)
    # detection score 
    # convert score to % to display in image
    score = "Score {}%".format(int(points['det_score']*100))
    # y2+10 to print 10 pixels below the box
    cv2.putText(image_v2,score,(x1,y1),cv2.FONT_HERSHEY_DUPLEX,0.5,(255,255,255))
    
    # add age and gender
    gender = gender_data[points['gender']]
    age = points['age']
    
    age_gender_text = f"{age} {gender}"
    cv2.putText(image_v2, age_gender_text,(x1,y2+20),cv2.FONT_HERSHEY_DUPLEX,0.5,(255,0,255))
  
resized_img = cv2.resize(image_v2, (1800, 924))
cv2.imshow('Detected faces',resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
