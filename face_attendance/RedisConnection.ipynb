```python
db_key = 'school:register'
# connect to redis database
hostname = 'redis-18992.c258.us-east-1-4.ec2.redns.redis-cloud.com'
port = '18992'
password = 'nUoPu0pKGDk95n5zatr1N1DgTDNmnAuG'
r = redis.StrictRedis(host=hostname,
                      port=port, password=password)
```

```python
def find_substring_in_hash(substring):
    cursor, data = r.hscan(db_key)
    result = {k.decode('utf-8'): v for k, v in data.items() if substring in k.decode('utf-8')}
    return result
```

```python
def check_login(email, passw):
    data = find_substring_in_hash(email)
    if len(data) < 1:
        return {'message': "User not found", 'data': {}}
    else:
        for key in data:
            name, role, emailid, password = key.split('#')

            if emailid == email and password == passw:
                return {'message': "Success", 'data': {"name": name, "role": role, "email": email}}

        return {'message': "Incorrect email or password", 'data': {}}
```
